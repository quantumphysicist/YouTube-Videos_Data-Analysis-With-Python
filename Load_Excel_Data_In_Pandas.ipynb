Because I explained each step, the above might have seem laborious. Here it all is in one code block:

```python
# Get the names of the sheets in the Excel file
sheet_names = pd.ExcelFile(excel_file).sheet_names
print(sheet_names)

# Read the data from the correct sheet
# Merge rows 1 and 2
df = pd.read_excel(excel_file, sheet_name="Annualised", skiprows = 5, header = [0,1])

# Make it a single level index
df.columns = [' '.join(col).strip() for col in df.columns.values]

# In df.columns, replace 'Unnamed /d+' with '' (The \d+ is a regular expression for a digit)
df.columns = df.columns.str.replace('Unnamed: \d+_level_0 ', '', regex=True)

# Convert the appropriate columns to integer data type
df[df.columns[6:14]] = df[df.columns[6:14]].astype('int32') 

df.head()
```
