```python
import pandas as pd
```

```python
# Usually to do web scraping,
# we would use the Python library called BeautifulSoup

# It turns out that pandas allows you to do some
# simple web scraping using the read_html method
```

```python
url = 'https://en.wikipedia.org/wiki/Central_line_(London_Underground)'
```

```python
list_of_dfs = pd.read_html(url)
```

```python
# We need to find the table with the stations (i.e. with the column 'Station')
# We can do this by looking at the columns by iterating through the list of dataframes:
for index, df in enumerate(list_of_dfs):
    print(index, df.columns)
```

```python
df = list_of_dfs[32]
```
